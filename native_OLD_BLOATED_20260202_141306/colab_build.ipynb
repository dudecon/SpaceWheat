# SpaceWheat GDExtension Build (Google Colab)

This notebook compiles the SpaceWheat native GDExtension on Google's servers.

**Steps:**
1. Upload `native_source.tar.gz` (created by `prepare_colab.sh`)
2. Run all cells
3. Download `libquantummatrix.linux.template_debug.x86_64.so`
4. Place it in `~/ws/SpaceWheat/native/bin/`

```python
# Install build tools
!apt-get update -qq
!apt-get install -y -qq build-essential scons python3-dev
```

```python
# Extract source (upload native_source.tar.gz first)
!tar -xzf native_source.tar.gz
!ls -lh native/
```

```python
# Build (uses all cores, ~2-3 minutes)
!cd native && scons -j$(nproc)
!ls -lh native/bin/*.so
```

```python
# Verify the build
!file native/bin/libquantummatrix.linux.template_debug.x86_64.so
!ldd native/bin/libquantummatrix.linux.template_debug.x86_64.so
```

## Download the compiled library

Right-click on `native/bin/libquantummatrix.linux.template_debug.x86_64.so` in the file browser and download it.

Then place it in `~/ws/SpaceWheat/native/bin/` on your local machine.
